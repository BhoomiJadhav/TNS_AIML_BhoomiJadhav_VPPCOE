```python
# for loop
for i in range(1, 6):
    print(i)

# while loop
i = 1
while i <= 5:
    print(i)
    i += 1

# nested for loop
for i in range(5):
    print("Hey")
    for j in range(6):
        print("Hi")
        for x in range(2):
            print("Hello")

# function definition
def conditional():
    if True:
        for i in range(5):
            print("Hello")
    else:
        print("Not hello")

# functions in arguments
def work(a, b):
    print(a)
    print(b)

work("hello", "world")

```
