## **A List is Python is a dynamic, ordered, mutable sequence of elements**

# Characteristics:
*   Indexed- Index are accessed via zero based index
*   Mutable- can Modify
*   Allow Duplicates
*   Heterogenous- can store different data types
*   Dynamic size


# Common Operations in Lists:

extend()  
insert()  
remove()  
pop()  
clear()  
index()  
count()  
reverse()

copy()  
len()

```python
#creating a list
list =[10,23,"a",51.5]
print("List: ",list)
```

# **1. Append**


```python
#add a single element to the end of list
list.append(40)
list.append("Bhoomi")
print("After Append List:" , list)
```

```python
car_list=["alto","kia","Rolls Royce"]
print("car list: ",car_list)
car_list.append("BMW")
print("After Append List:" , car_list)
car_list.append("Audi")
print("After Append List:" , car_list)
```

# **2.extend**

```python
numbers1=[10,20,30]
numbers2=[40,50,60,70]
numbers1.extend(numbers2)
print("After Extend List:" , numbers1)
```

```python
#add all elements from another iterable(like another list)
list.extend(car_list)
print("After Extend List:" , list)
```

# **3.Insert**

```python
#in insert we can inser a element at any position
subject=["DSA","Python","Java"]
subject.insert(1,"C")
subject.insert(3,"java")
#subject.insert(1,"C++",2,"JS")
print("after Inserting: ",subject)
```

# **4.Remove**

```python
subject.remove("C")
subject.remove("java")
print("after removing: ",subject)
```

# **5.Pop**

```python
subject.pop()
print("after poping: ",subject)
```

```python
subject.pop(0)
print("after poping: ",subject)
```

```python
subject.extend(["Java","C","C++","JS"])
```

# **6.Index**

```python
#gives first occurence of element index value
subject.index("JS")
```

## **7.Count**

```python
subject.count("C")
```

```python
subject.reverse()
print("Reverse List: ",subject)
```

```python
subject.sort()
print("Sorted List: ",subject)
numbers1.sort(reverse=True)
print("Sorted List in Descending Order: ",numbers1)
```
