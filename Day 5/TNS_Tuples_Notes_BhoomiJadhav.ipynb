**A Tuple in python is a ordered,immutable sequence of element**

**Characteristics**:
1. Indexed
2. Immutable
3. Allow Duplicates
4. Heterogenous
5. Faster than Lists

**Implementation**:-
Tuples are implemented as static arryas, when the size is fixed after creation. This makes them faster and more memory-efficient and read only size

**Common Operations:-**  
index()  
count()  
max()/men()  
len()  
sum()  
tuple()  

```python
my_tuple=(1,2,2,"hii",'a',3,4,5)
print("Tuple: ", my_tuple)
```

```python
print("index of 2:", my_tuple.index(2))
print("count of 4:",my_tuple.count(4))

print("len of tuple:",len(my_tuple))

```

```python
example=(1,2,33,4,5,66,66,7)
print("max ele: ",max(example))
print("min ele: ",min(example))
print("sum of tuple: ",sum(example))
```
